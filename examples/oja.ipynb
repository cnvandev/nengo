```python
import numpy as np

import nengo

dt = 0.001
n = 500
n_dope = 50
d = 4
intercepts = nengo.objects.Uniform(0.1, 1.0)

def keys(t):
    if t < 0.5:
        return [1.0, 0, 0, 0]
    if t < 1:
        return [0, 1.0, 0, 0]
    if t < 1.5:
        return [0.0, -1.0, 0, 0]
    if t < 2.0:
        return [0.7, 0, -0.7, 0]
    def k(t):
        if t < 2.5:
            return [1.0, 0, 0, 0]
        if t < 3.0:
            return [0, 1.0, 0, 0]
        if t < 3.5:
            return [0.0, -1.0, 0, 0]
        if t < 4.0:
            return [0.7, 0, -0.7, 0]
    return np.asarray(k(t)) + np.random.normal(0, 0.1, 4)

def values(t):
    if t < 0.5:
        return [0.0, -1.0, 0, 0]
    if t < 1.0:
        return [0.7, 0, -0.7, 0]
    if t < 1.5:
        return [1.0, 0, 0, 0]
    if t < 2.0:
        return [0.0, -1.0, 0, 0]
    return [0, 0, 0, 0]

def is_learning(t):
    return [int(t < 2.0)]

m = nengo.Model()
key = nengo.Node(output=keys)
value = nengo.Node(output=values)
learning = nengo.Node(output=is_learning)
bias = nengo.Node(output=[1])
dopamine = nengo.Ensemble(nengo.LIF(n_dope), 1, intercepts=intercepts, encoders=[[1]]*n_dope)
antidopamine = nengo.Ensemble(nengo.LIF(n_dope), 1, intercepts=intercepts, encoders=[[1]]*n_dope)
x = nengo.Ensemble(nengo.LIF(n), d, label="x", intercepts=nengo.objects.Uniform(-0.1, 1.0))
y = nengo.Ensemble(nengo.LIF(n), d, label="y", intercepts=intercepts)
y_est = nengo.Ensemble(nengo.LIF(n), d, label="y_est")
e = nengo.Ensemble(nengo.LIF(n), d, label="error")

def inhibit(pre, post, amount=20, filter=0.001):
    nengo.Connection(pre.neurons, post.neurons, transform=-amount*np.ones((post.n_neurons, pre.n_neurons)), filter=filter)

nengo.Connection(key, x)
nengo.Connection(x.neurons, y.neurons, transform=2.0*np.eye(n),
    learning_rule=nengo.OJA(learning_rate=1e-4, learning=dopamine, scale=2.0))
nengo.Connection(value, e)
nengo.Connection(y_est, e, transform=-1)
nengo.Connection(learning, dopamine)
nengo.Connection(bias, antidopamine)
inhibit(dopamine, antidopamine)
inhibit(antidopamine, e)
nengo.Connection(y, y_est,
    learning_rule=nengo.PES(e, learning_rate=1e-3))

p_key = nengo.Probe(key, "output", filter=0.03)
p_value = nengo.Probe(value, "output", filter=0.03)
p_x = nengo.Probe(x, "decoded_output", filter=0.03)
spikes_x = nengo.Probe(x, "spikes")
p_y = nengo.Probe(y, "decoded_output", filter=0.03)
spikes_y = nengo.Probe(y, "spikes")
p_e = nengo.Probe(e, "decoded_output", filter=0.03)
p_dope = nengo.Probe(dopamine, "decoded_output", filter=0.03)
p_antidope = nengo.Probe(antidopamine, "decoded_output", filter=0.03)
p_y_est = nengo.Probe(y_est, "decoded_output", filter=0.03)
#spikes_y_est = nengo.Probe(y_est, "spikes")
```

```python
from nengo.matplotlib import rasterplot

sim = nengo.Simulator(m, dt)
sim.run(4)

figure()
title('Key')
plot(sim.data(p_key))

figure()
title('Value')
plot(sim.data(p_value))

figure()
title('x Spikes')
rasterplot(sim.trange(), sim.data(spikes_x))

figure()
title('Memory')
plot(sim.data(p_y))

figure()
title('Memory Spikes')
rasterplot(sim.trange(), sim.data(spikes_y))

figure()
title('y')
plot(sim.data(p_y_est))

#figure()
#title('y Spikes')
#rasterplot(sim.trange(), sim.data(spikes_y_est))

figure()
title('Error')
plot(sim.data(p_e), label='Error')
plot(sim.data(p_dope), label='Dopamine')
plot(sim.data(p_antidope), label='No Learning')
legend(loc='best')
show()
```
